```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mobile=['samsung','apple','Redmi','Vivo','Oppo']
online=np.random.randint(10000,20000,5) #(low,high,size(how many random integers we want))
offline=np.random.randint(10000,20000,5)

plt.title("showing graph of online and offline selling")
plt.plot(mobile,online,label='online',linestyle="--",color="k",marker="p",markersize="9")
plt.plot(mobile,offline,label='offline',linestyle="--",color="g",marker="*",markersize="9")
plt.xticks(rotation=70)
plt.legend()
plt.show()

```

```python
a=np.arange(1,11)
b=np.arange(11,21)
print(a)
print(b)

plt.figure(figsize=(4,5))
plt.title("this is my linear graph")
plt.plot(a,b,color='silver',marker='h',markersize=9,linestyle='--',linewidth=6)
plt.plot(a,a**2,color='gold',marker='h',markersize=9,linestyle='--',linewidth=6)
plt.xticks(a) # for taking all step size of x-axis and y-axis
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mobile=['samsung','apple','Redmi','Vivo','Oppo']
online=np.random.randint(10000,20000,5) #(low,high,size(how many random integers we want))
offline=np.random.randint(10000,20000,5)

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(mobile,online,label='online',linestyle="--",color="k",marker="p",markersize="9")
plt.legend()
plt.subplot(1,2,2)
plt.plot(mobile,offline,label='offline',linestyle="--",color="g",marker="*",markersize="9")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.subplot(2,1,1)
plt.plot(mobile,online,label='online',linestyle="--",color="k",marker="p",markersize="9")
plt.legend()
plt.subplot(2,1,2)
plt.plot(mobile,offline,label='offline',linestyle="--",color="g",marker="*",markersize="9")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mobile=['samsung','apple','Redmi','Vivo','Oppo']
online=np.random.randint(10000,20000,5) #(low,high,size(how many random integers we want))
offline=np.random.randint(10000,20000,5)

plt.title("showing graph of online and offline selling")
plt.xlabel("MOBILE PHONES")
plt.ylabel("SELLING")
plt.scatter(mobile,online,label='online',linestyle="--",color="k",marker="p",s=250)
plt.scatter(mobile,offline,label='offline',linestyle="--",color="g",marker="*",s=250)
plt.xticks(rotation=70)
plt.legend()
plt.show()
```

```python
# pie chart - It is used to find the proportion from the data
```

```python
import numpy as np
y=np.array([35,20,15,30])
fr=['apple','mango','banana','grapes']
L=[0,0,0,0.1]
plt.pie(y,labels=fr,explode=L,startangle=90,autopct="%0.2f%% ")
plt.show()
```
